```python
import networkx as nx

G = nx.read_gml('./grafo_con_poda.gml') #dirigido
```

```python
G = G.to_undirected()
```

```python
# Paso 4: Realizar análisis adicional en el grafo (opcional)
print("Número de nodos:", G.number_of_nodes())
print("Número de aristas:", G.number_of_edges())

```

```python
nx.is_connected(G) #es conecto porque es no dirigido
```

```python
grados = dict(G.degree()) 

#for nodo, grado in grados.items():
#  print(f"El {nodo} tiene {grado} aristas")

mas_sociable = max(grados.items(), key = lambda nodo: nodo[1]) #Cuando usas grados, estás accediendo solo a los valores (los grados de los nodos), mientras que cuando usas grados.items(), estás accediendo tanto a las claves como a los valores (los nodos y sus grados).
menos_sociable = min(grados.items(), key = lambda nodo: nodo[1])

print("El usuario más sociable es:", mas_sociable)
print("El usuario menos sociable es:", menos_sociable)
```

```python
import seaborn as sns

sns.set_theme(style="whitegrid") #barras con fondo blanco y lineas en gris claro

grados = [d for n, d in G.degree()] #extrae los grados "d" por cada "n"(nodo)

sns.countplot(x=grados)
```

```python
grados_ord = sorted(set(grados)) #set convierte la lista "grados" en un conjunto eliminando valores repetidos, 

conteo = [grados.count(g) for g in grados_ord] #itera sobre cada grado "g" en grados_ord y cuenta cuántas veces aparece, es decir, frecuencia de cada grado

f, ax = plt.subplots(figsize=(7, 7)) #crea una figura de conjunto de ejes estableciendo el tamaño en pulagadas. f=figura, ax=eje/s
ax.set(xscale="log", yscale="log") #establece la escala de los ejes, logarítmica
sns.scatterplot(x=grados_ord, y=conteo, ax=ax) #diagrama de dispersión 
```

```python
# Aplicar un algoritmo de detección de comunidades (por ejemplo, Louvain)
comunidades = nx.algorithms.community.modularity_max.greedy_modularity_communities(G) #el algoritmo de louvain encuentra comunidades maximizando la modularidad

# Mostrar el número de comunidades identificadas
print("Número de comunidades encontradas:", len(comunidades))

# Mostrar los nodos en cada comunidad
for i, comunidad in enumerate(comunidades):
    print(f"Comunidad {i+1}: {len(comunidad)} nodos")
    print("Nodos:")
    for node in comunidad:
        print(node)
    print() 

```
